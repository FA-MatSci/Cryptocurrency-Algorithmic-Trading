```python
import pandas as pd
import numpy as np
import sqlalchemy     #for SQL Database
from sqlalchemy import create_engine
from binance import Client
from binance import BinanceSocketManager   #to interact with websocket

import asyncio
from binance import AsyncClient, BinanceSocketManager
import nest_asyncio

import time
import pymysql

import json  #to transform data from API
import websocket   #library used to connect to websocket API

import smtplib   #to send yourself an e-mail with recommendations

import datetime

import math
from datetime import datetime, timedelta
import time
import concurrent.futures

import warnings

from IPython.core.display import display_html
import os
```

```python
def fun(coin):
    API_KEY = 'xxx'
    API_SECRET = 'yyy'
    client = Client(API_KEY, API_SECRET)
    pymysql.install_as_MySQLdb() 
    nest_asyncio.apply()
    
    engine = sqlalchemy.create_engine('mysql://root:password@localhost:number/')

    with engine.connect() as connection:
        connection.execute(sqlalchemy.schema.CreateSchema(coin))

    def createframe(msg):
        price = msg['k']['c']
        price = float(price)
        time = msg['E']
        time = pd.to_datetime(time, unit='ms')
        df = pd.DataFrame(price, columns=[msg['s']], index=[time])
        df = df.reset_index()
        df = df.rename(columns={'index':'Timestamp'})
        return df


#    here you get price data every 2 seconds!! 

    async def main(coin):
        try:
            client = await AsyncClient.create()
            bm = BinanceSocketManager(client)
            ks = bm.kline_socket(coin)
            engine = sqlalchemy.create_engine('mysql://root:password@localhost:3306/{}'.format(coin))
            async with ks as tscm:
                while True:
                    try:
                        res = await tscm.recv()
                        frame = createframe(res)     #for data cleaning
#                        print(frame)
                        frame.to_sql(coin, engine, if_exists='append', index=False)     #my approach, you send data into sql database (MySQLWorkbench). Here you create a table
                        with open("status.txt", "r") as f:                          #beforehand, create txt file
                            status = f.read()
                            if status == "stopped":
                                display_html("<script>Jupyter.notebook.kernel.restart()</script>", raw=True)   #kernel restarts
                            else:
                                pass

                    except:
                        print('A connection loss occured!!')
                        if __name__ == "__main__":
                            loop = asyncio.get_event_loop()
                            loop.run_until_complete(main(coin))


        except:
            print('There are problems with the connection!! We try to connect again in 30 sec.')
            time.sleep(30)
            if __name__ == "__main__":
                loop = asyncio.get_event_loop()
                loop.run_until_complete(main(coin))


    if __name__ == "__main__":
        loop = asyncio.get_event_loop()
        loop.run_until_complete(main(coin))
```
